```python
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
from Bio.SeqFeature import SeqFeature, FeatureLocation
```

```python
for record in SeqIO.parse("incompleto_genomic_mtb.gbff", "genbank"):
  print(record)
  print(record.id)
```

```python
for record in SeqIO.parse("incompleto_genomic_mtb.gbff", "genbank"):
  for feature in record.features:
    print(feature)
```

```python
#Abrindo o arquivo e lendo a ficha do genoma - Anotaes gerais
for record in SeqIO.parse("incompleto_genomic_mtb.gbff", "genbank"):
  for feature in record.features:
    if feature.type == "CDS":
      if 'gene' in feature.qualifiers:
        print(feature.qualifiers['gene'])
```

```python
for record in SeqIO.parse("incompleto_genomic_mtb.gbff", "genbank"):
  for feature in record.features:
    if feature.type == "CDS":
      if 'gene' in feature.qualifiers:
        if 'gyrB' in feature.qualifiers['gene']:   #Procurando a localizao do gene gyrB no qualifier "gene", e este se encontra na fita senso
          print(record.id, feature.location)
          print(record,id, feature.qualifiers['translation'])  #translation  o qualifier da sequncia de aminocidos
#O SINAL (+) SIGNIFICA QUE O GENE EST NA FITA SENSO
```

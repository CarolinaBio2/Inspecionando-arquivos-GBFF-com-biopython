```python
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
from Bio.SeqFeature import SeqFeature, FeatureLocation
```

```python
#Abrindo o arquivo e lendo a ficha do genoma + anotações gerais
record = SeqIO.read("completo_genomic_ecoli.gbff", "genbank")
print(record.annotations)
print(list(record.annotations))
print(record.annotations['organism'])
print(record.annotations['references'])
print(record.dbxrefs)
```

```python
print(record.letter_annotations) #anotações secundárias sobre estrutura, tamanho da sequência ou qualquer outra nota
print(record.features) #Começando a acessar as features
print(record.features[0]) #A primeira é sempre a source
print(record.features[0].type)
```

```python
#Acessando os qualifiers(qualificadores da sequência) de cada feature(cada característica do arquivo e da sequência)
print(record.features[1].qualifiers)
print(record.features[0].qualifiers['organism'])  #para organismo o único qualifier possível é 0
print(record.features[5].qualifiers)
print(record.features[2].qualifiers['gene'])
print(record.features[4].qualifiers)
print(record.features[10].qualifiers['product'])
print(record.features[2].qualifiers['translation'])
```

```python
#COMO ITERAR SORBE A LOCALIZAÇÃO E ACESSAR A SEQUÊNCIA DOS GENES; lembrando que no python a primeira posição é 0
print(record.features[5].location)  #colocando números aleatórios para os intervalos de localização
print(record.features[8].strand) #fita senso é 1 (+) e antisenso é -1 (-)
print(record.features[25].location.start) #localização no início
print(record.features[85].location.end)  #localização no final
```

```python
#Lendo a record
record = SeqIO.read("completo_genomic_ecoli.gbff", "genbank")

#Recuperando o start e o end da feature de interesse
start = record.features[4].location.start
end = record.features[8].location.end

#Acessando a sequência do cromossomo todo
myseq = record.seq

#Criando as coordenadas e extraindo a sequência de interesse
coord = SeqFeature(FeatureLocation(start,end))
gene_interesse = coord.extract(myseq)

gene_id = record.features[2].qualifiers['gene']

print(gene_id , gene_interesse)
```

```python
#Recuperando o start e o end da feature de interesse
start = record.features[2].location.start
end = record.features[8].location.start

#Acessando a sequência do cromossomo todo
myseq = record.seq

#Criando as coordenadas e extraindo a sequência de interesse
coord = SeqFeature(FeatureLocation(start,end)) #localização do começo ao fim do gene
gene_interesse = coord.extract(myseq)

if 'gene' in record.features[8].qualifiers:
	gene_id = record.features[8].qualifiers['gene'] #definindo uma variável para armazenar o id do gene (caracteristica da posição 9 da record, dentro do qualificador gene)
	print(gene_id , gene_interesse)

else:
	print(gene_interesse)
```
